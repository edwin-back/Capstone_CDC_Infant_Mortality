```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
AGE_mortality = pd.read_csv ('Mortality rates AGE birthplace.csv')
```

```python
AGE_mortality
```

```python
print(AGE_mortality.dtypes.to_string())
```

```python
AGE_mortality.hist(column= 'Deaths')
```

```python
AGE_mortality.hist(column= 'Births')
```

```python
state_birth = AGE_mortality.groupby(['State', 'Deaths']).mean()
state_birth
```

```python
deaths_age= AGE_mortality[['State', 'Deaths']]
deaths_age
```

```python
#deaths_age.groupby(['State', 'Deaths'])
Grouped_deaths_state= deaths_age.groupby("State")[["Deaths"]].mean()
Grouped_deaths_state
```

```python
Grouped_deaths_state['State'] = Grouped_deaths_state.index
Grouped_deaths_state.index
Grouped_deaths_state
```

```python
print(Grouped_deaths_state.dtypes.to_string())
```

```python
Deaths_State= Grouped_deaths_state.plot.bar(x="State", y="Deaths", title="Deaths by State", figsize=(20,10), color='#86bf91')
```

```python
births_age= AGE_mortality[['State', 'Births']]
births_age
```

```python
Grouped_births_state= births_age.groupby("State")[["Births"]].mean()
Grouped_births_state
```

```python
Grouped_births_state= births_age.groupby("State")[["Births"]].mean()
Grouped_births_state
Grouped_births_state['State'] = Grouped_births_state.index
Grouped_births_state.index
Grouped_births_state
```

```python
Births_State= Grouped_births_state.plot.bar(x="State", y="Births", title="Births by State", figsize=(20,10), color='#86bf91')
```

```python
State_Birth_Death_Avg= pd.merge(Grouped_births_state, Grouped_deaths_state, on='State')
State_Birth_Death_Avg
```

```python
State_Birth_Death_Avg['Result'] = State_Birth_Death_Avg['Deaths']/State_Birth_Death_Avg['Births']
State_Birth_Death_Avg[['Result']]
death_rates_state= State_Birth_Death_Avg[['Result']]
death_rates_state
```

```python
death_rates_state['State'] = death_rates_state.index
death_rates_state.index
death_rates_state
Death_rates_bar= death_rates_state.plot.bar(x="State", y="Result", title="Death Rates by State", figsize=(20,10), color='#86bf91')
```
